Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,

1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('iris')
dataset
```

```python
print("Numeric : ", dataset.select_dtypes(include=['number']).columns.tolist())
print("Nominal : ", dataset.select_dtypes(include=['object']).columns.tolist())
```

```python
fig, axis = plt.subplots(2, 2, figsize=(16, 9))
sns.histplot(dataset["sepal_length"], ax =axis[0,0], color="r", kde=1)
sns.histplot(dataset["sepal_width"], ax =axis[0,1], color="g", kde=1)
sns.histplot(dataset["petal_length"], ax =axis[1,0], color="b", kde=1)
sns.histplot(dataset["petal_width"], ax =axis[1,1], color="y", kde=1)
```

1. Sepal length is evenly distributed
2. Speal width is normally distributed
3. Petal length is right skewed 
4. Petal width is right skewed

```python
fig , axes = plt.subplots(2, 2, figsize=(16, 9))
sns.boxplot(x="species", y="sepal_length", data=dataset, ax = axes[0,0], color='r')
sns.boxplot(x="species", y="sepal_width", data=dataset, ax = axes[0,1], color='g')
sns.boxplot(x="species", y="petal_length", data=dataset, ax = axes[1,0], color='b')
sns.boxplot(x="species", y="petal_width", data=dataset, ax = axes[1,1], color='y')
```
