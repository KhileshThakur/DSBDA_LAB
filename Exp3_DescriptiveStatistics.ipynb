### 1.  Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable.

```python
import numpy as np
import pandas as pd
import random 
```

```python
stud_id = range(1, 101)
stud_name = ["stud "+ str(i) for i in stud_id]
income = [random.randint(50000, 300000) for i in stud_id]
age_group = [random.choice(["NewBorn", "Elder", "Young"]) for i in stud_id]

dataset = pd.DataFrame({
            "Stud Id" : stud_id,
            "Stud Name" : stud_name,
            "Income" : income,
            "Age Group" : age_group
})
dataset
```

```python
dataset.groupby('Age Group').describe()
```

```python
dataset.groupby("Age Group")["Income"].describe()
```

```python
dataset.loc[5:10, "Income"].mean() #just for knowledge
```

```python
dataset.groupby("Age Group")["Income"].mean()
```

```python
dataset.groupby("Age Group")["Income"].median()
```

```python
dataset["Income"].mode()
```

```python
dataset.groupby("Age Group")["Income"].min()
```

```python
dataset.groupby("Age Group")["Income"].max()
```

```python
dataset.groupby("Age Group")["Income"].std()
```

```python
dataset.groupby("Age Group")["Income"].var()
```
